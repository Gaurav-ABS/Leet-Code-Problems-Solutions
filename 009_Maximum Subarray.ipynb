# MaximumSubarray

- Problem Description on Leetcode

  <a href ="https://leetcode.com/problems/maximum-subarray/" > Maximum Subarray </a>

- Method:

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]

print (len(nums))

window=1

subarrays = list()


while window <= len(nums):
    
    for startPosition in range(len(nums)-(window)+1):
        subarrays.append(nums[startPosition:startPosition+window])
        
    print('Window ===' ,window)
    
    print(subarrays)
    
    print('_________________________')
    
    window = window + 1
    
    
sumOfSubarrays=list()
    
for k in range(len(subarrays)):
    sumOfSubarrays.append(sum(subarrays[k]))
print('List containing sum of the subarrays ==== ',sumOfSubarrays)

print("_________________FINAL RESULT________________________")

print('The maximum sum possible ====',max(sumOfSubarrays))

print('The array with the largest sum ====',subarrays[sumOfSubarrays.index(max(sumOfSubarrays))])


    
    
    
    
```
